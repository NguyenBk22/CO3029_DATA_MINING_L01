```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules, fpgrowth
import time

# Tải dữ liệu đã làm sạch từ file Data_processing.ipynb
basket_sets = pd.read_csv("..\cleaned_basket.csv", index_col=0)
```

```python
    
# 1. ĐỊNH NGHĨA CÁC NGƯỠNG CẦN THỬ
# Chúng ta sẽ thử 2%, 1.5%, 1%, và 0.5%
min_support_levels = [0.02, 0.015, 0.01, 0.005]
    
results = []
total_transactions = len(basket_sets)

```

```python
# 2. CHẠY VÒNG LẶP THỬ NGHIỆM
print(f"Bắt đầu thử nghiệm trên {total_transactions} giao dịch...")
for min_support in min_support_levels:
    start_time = time.time()
    
    # Tính số giao dịch tối thiểu
    absolute_support = int(min_support * total_transactions)
        
    # Chạy Apriori
    frequent_itemsets = apriori(basket_sets, 
                            min_support=min_support, 
                            use_colnames=True, 
                            low_memory=True)
    num_itemsets = len(frequent_itemsets)
        
    num_rules = 0
    if num_itemsets > 0:
        # Sinh luật
        rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
        # Lọc các luật có ý nghĩa (ví dụ: confidence > 0.5)
        rules_filtered = rules[(rules['confidence'] >= 0.5)]
        num_rules = len(rules_filtered)
            
    end_time = time.time()
        
        # Lưu kết quả
    results.append({
            "Min Support (%)": f"{min_support*100:.1f}%",
            "Số giao dịch tối thiểu": absolute_support,
            "Số tập phổ biến (Itemsets)": num_itemsets,
            "Số luật (Rules)": num_rules,
            "Thời gian (giây)": f"{end_time - start_time:.2f}s"
})

  
```

```python
# 3. IN BẢNG TÓM TẮT KẾT QUẢ
results_df = pd.DataFrame(results)
print("\n--- KẾT QUẢ THỬ NGHIỆM ---")
print(results_df.to_string())
```
